```python
with open('documents.json', 'r', encoding='utf-8') as f:
    documents_raw = json.load(f)
```

```python
documents_raw
```

```python
documents = []
for topic_data in documents_raw:
    topic = topic_data['topic']
    for doc in topic_data['documents']:
        doc['topic'] = topic
        documents.append(doc)
```

```python
documents
```

```python
df = pd.DataFrame(documents, columns=['topic', 'timestamp', 'title', 'content'])
```

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import _VectorizerMixin

analyzer = CountVectorizer().build_analyzer()
indonesian_stopwords = stopwords.words('indonesian')

processed_stopwords = set()
for word in indonesian_stopwords:
    tokens = analyzer(word)
    processed_stopwords.update(tokens)
```

```python
cv = CountVectorizer(min_df=3, stop_words=list(processed_stopwords))
```

```python
cv.fit(df.content)
```

```python
cv.get_feature_names_out()
```

```python
cv.get_feature_names_out().shape
```

```python
MATRIX = cv.fit_transform(df.content)
names = cv.get_feature_names_out()
df_docs = pd.DataFrame(MATRIX.toarray(), columns=names).T
df_docs
```
